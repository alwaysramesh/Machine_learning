```python
import pandas as pd 
```

```python
dataset=pd.read_csv("C://Users//ALWAYSRAMESH//Downloads//insurance.csv")
```

```python
dataset.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
model=LabelEncoder()
dataset['sex']=model.fit_transform(dataset['sex'])
dataset['smoker']=model.fit_transform(dataset['smoker'])
```

```python
x=dataset.iloc[:,:-2]
```

```python
x
```

```python
y=dataset['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
modelknn= KNeighborsRegressor(n_neighbors=1) 
modelknn.fit(x_train,y_train)
```

```python
y_pred=modelknn.predict(x_test)
```

```python
y_pred2=modelknn.predict([[19,0,27.900,0,1]])
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,accuracy_score
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
print(r2_score(y_test,y_pred))
```
