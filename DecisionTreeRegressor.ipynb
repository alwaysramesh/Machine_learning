```python
import pandas as pd 
```

```python
dataset=pd.read_csv("C://Users//ALWAYSRAMESH//Downloads//insurance.csv")
```

```python
dataset.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
model=LabelEncoder()
dataset['sex']=model.fit_transform(dataset['sex'])
```

```python
dataset['smoker']=model.fit_transform(dataset['smoker'])
```

```python
x=dataset.iloc[:,:-2]
y=dataset['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model2=DecisionTreeRegressor(max_leaf_nodes=5,max_depth=3)
model2.fit(x_train,y_train)
```

```python
y_pred=model2.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt 
print(plot_tree(model2))
plt.show()
```
