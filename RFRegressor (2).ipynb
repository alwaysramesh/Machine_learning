```python
import pandas as pd 
```

```python
dataset=pd.read_csv("C://Users//ALWAYSRAMESH//Downloads//insurance.csv")
```

```python
dataset.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
model=LabelEncoder()
dataset['sex']=model.fit_transform(dataset['sex'])
```

```python
dataset['smoker']=model.fit_transform(dataset['smoker'])
```

```python
x=dataset.iloc[:,:-2]
y=dataset['charges']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=30)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor(min_samples_split=2)
model2.fit(x_train,y_train)
```

```python
print(x_train.shape) 
print(y_train.shape)
```

```python
y_pred=model2.predict(x_test)
```

```python
#y_pred=model2.predict([['19','0','27.900','0','1']])
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test,y_pred))
```
