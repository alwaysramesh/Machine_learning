```python
import pandas as pd 
```

```python
dataset=pd.read_csv("C://Users//ALWAYSRAMESH//Downloads//insurance.csv")
```

```python
dataset.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
dataset['sex']=le.fit_transform(dataset['sex'])
#dataset['sex'=le.fit_transform(['sex'])]
```

```python
x=dataset.iloc[:,:-3]
```

```python
y=dataset['smoker']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
#from sklearn.neighbors import KNeighborsClassifier
#knn = KNeighborsClassifier(n_neighbors=3)
```

```python
model=AdaBoostClassifier(n_estimators=50,learning_rate=0.01)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt 
from sklearn.tree import plot_tree
plot_tree(model.estimators_[40],filled=True)
plt.show()
# estimators_ me [2] kon si tree leni hai uski value underscore me dena padega bagging and boosting me
```
