```python
import pandas as pd 
import numpy as np 
```

```python
dataset=pd.read_csv("C://Users//ALWAYSRAMESH//Downloads//insurance.csv")
```

```python
dataset.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
dataset['sex']=le.fit_transform(dataset['sex'])
dataset['smoker']=le.fit_transform(dataset['smoker'])
#dataset['region']=le.fit_transform(dataset['region'])
```

```python
x=dataset.iloc[:,:-2]
```

```python
y=dataset['region']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
li=RandomForestClassifier()
```

```python
forest_params=[{'n_estimators':[50,75,100],'max_depth':list(range(10,15)),'max_features':list(['sqrt','log2'])}]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.model_selection import GridSearchCV
model=RandomizedSearchCV(estimator=li, param_distributions=forest_params, cv=5, scoring='accuracy')
model.fit(x_train,y_train)
```

```python
model.fit(x_train, y_train)
print(model.best_params_)
print(model.best_score_)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
