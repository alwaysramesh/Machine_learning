```python
import pandas as pd
```

```python
df=pd.read_csv("C://Users//ALWAYSRAMESH//Downloads//insurance.csv")
```

```python
df
```

```python
x=df[['age','bmi','children','charges']]
```

```python
y=df['sex']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(penalty='l1',solver='saga',C=0.005) 
model.fit(x_train, y_train) 
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, y_pred))
```

```python
cm=confusion_matrix(y_test, y_pred)
# Here Multyclass so 4*4 matrix
print(cm)
```

```python
print(classification_report(y_test,y_test))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred 0', 'Pred 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix Heatmap")
plt.show()
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='summer', xticklabels=['Pred 0', 'Pred 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix Heatmap")
plt.show()
```
