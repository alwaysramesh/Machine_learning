```python
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report
import seaborn as sns
```

```python
df = pd.read_csv("C://Users//ALWAYSRAMESH//Downloads//spam.csv", encoding='latin1')
```

```python
df.head(2)
```

```python
x=df['v2']
y=df['v1']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_clf=CountVectorizer(stop_words='english',ngram_range=(3,4))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
x_train1=bow_clf.fit_transform(x_train)
```

```python
x_train.head(20)
```

```python
bow=CountVectorizer()
```

```python
x_train1=bow.fit_transform(x_train)
```

```python
x_test=bow.transform(x_test)
```

```python
model=MultinomialNB()
model.fit(x_train1,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
