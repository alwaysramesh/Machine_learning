```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split as tits
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\Gourav\\Downloads\\WELFake_Dataset.csv")
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
x = df['title']
y = df['label']
```

```python
x
```

```python
y
```

```python
bow = CountVectorizer()
```

```python
x = bow.fit_transform(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split = tits(x,y,test_size=0.3,random_state=42)
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_pred,y_test))
```
